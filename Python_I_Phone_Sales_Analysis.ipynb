```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

# **Exporting Dataset of Apple products**

```python
data=pd.read_csv("/apple_products.csv")
print(data.head())
```

```python
print(data.describe())
```

# **I Phone Sales Analysis in India**

```python
highest_rating=data.sort_values(by=["Star Rating"],ascending=False)
highest_rating=highest_rating.head(10)
print(highest_rating["Product Name"])
```

# **The number of ratings of the highest rated I phone on Flipkart**

```python
iphones= highest_rating["Product Name"].value_counts()
labels = iphones.index
counts = highest_rating["Number Of Ratings"]
fig = px.bar(highest_rating, x= labels, y= counts, title= "Number of ratings of highest rated I Phones")
fig.update_yaxes(range=[0,6000])
fig.show()
```

# **The number of reviews of highest reviewed I phone on Flipkart**

```python
iphones= highest_rating["Product Name"].value_counts()
labels = iphones.index
counts = highest_rating["Number Of Reviews"]
fig = px.bar(highest_rating, x= labels, y= counts, title= "Number of reviews of highest reviewed I Phones")
fig.update_yaxes(range=[0,600])
fig.show()
```

# **Number of Ratings VS Sale Price**

```python
figure=px.scatter(data_frame=data, x="Number Of Ratings", y="Sale Price",
                  size="Discount Percentage", trendline="ols",
                  title="Relationship between Sale prize and Number of ratings")
figure.show()
```

# **Discount Percentage VS Number of Ratings**

```python
figure=px.scatter(data_frame=data, x="Number Of Ratings", y="Discount Percentage",
                  size="Sale Price", trendline="ols",
                  title="Relationship between Discount Percentage and Number of Ratings")
figure.show()
```
