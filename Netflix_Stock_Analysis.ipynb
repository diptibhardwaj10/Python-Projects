# **Netflix Stock Analysis Project**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from datetime import datetime
```

```python
df=pd.read_csv("/netflix.csv")
```

```python
df.head(10)
```

# **Volume of Stock Traded**

```python
sns.set(rc={'figure.figsize':(10,5)})
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df=df.set_index("Date")
df.head()
```

```python
sns.lineplot(x=df.index,y=df["Volume"],label="Volume")
plt.title("Volume of Netflix Stock Traded")
plt.show()
```

# **Netflix Stock Prize**

```python
df.plot(y=["High","Close","Open"],title="Netflix Stock Prize(High,Close,Open) ")
```

```python
fig,(ax1,ax2,ax3) = plt.subplots(3,figsize=(15,10))
df.groupby(df.index.day).mean().plot( y="Volume",ax=ax1,xlabel = "Day", title="Netflix Stock Prize(Day)")
df.groupby(df.index.month).mean().plot(y="Volume",ax=ax2,xlabel="Month",title="Netflix Stock Prize(Month)")
df.groupby(df.index.year).mean().plot(y="Volume",ax=ax3,xlabel="Year",title="Netflix Stock Prize(Year)")
```

# **Highest & Lowest Stock Prize**

```python
a=df.sort_values("High",ascending=False).head(5)
a["High"]
```

```python
b=df.sort_values("Low",ascending=True).head(5)
b["Low"]
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,sharex=True, figsize=(12,5))
fig.suptitle("High & Low Stock Prize")
sns.lineplot(ax=axes[0],x=df.index,y=df["High"],label="High",color="green")
sns.lineplot(ax=axes[1],x=df.index,y=df["Low"],label="Low",color="red")
```
